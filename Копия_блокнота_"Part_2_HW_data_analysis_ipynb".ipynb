```
import datetime
# datetime (год, месяц, день)

def days(d1, d2):
    return d1- d2

d1 = datetime.datetime(2014, 7, 8)  
d2 = datetime.datetime(1980, 12, 12) 

print('Дней', days(d1, d2))
```

```
def months(d1, d2):
    return d1.month - d2.month + 12*(d1.year - d2.year)

d1 = datetime.datetime(2014, 7, 8)  
d2 = datetime.datetime(1980, 12, 12) 

print('Месяцев', months(d1, d2))
```

```
# Adj Close - отрегулированная цена закрытия
df1 = df1.rename(columns={"Adj Close":"AdjClose"}) # Переименовываем столбец иначе выдаст ошибку на графике
plt.figure(figsize=(13.5, 9))
plt.plot(df1.AdjClose, alpha=0.5, label='Adj Close - отрегулированная цена закрытия')
plt.xlabel('Годы')
plt.ylabel('Цена')
plt.legend(loc='upper center');
```
